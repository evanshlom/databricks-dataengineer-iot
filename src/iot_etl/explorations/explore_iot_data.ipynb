# IoT Data Exploration

Ad-hoc notebook for profiling the `/databricks-datasets/iot/iot_devices.json` dataset.
This notebook is NOT part of the SDP pipeline.

```python
iot = spark.read.json("/databricks-datasets/iot/iot_devices.json")
display(iot.limit(5))
```

```python
iot.printSchema()
```

```python
print(f"Total rows: {iot.count()}")
print(f"Null latitudes: {iot.filter(iot.latitude.isNull()).count()}")
print(f"Distinct countries: {iot.select("cn").distinct().count()}")
```
